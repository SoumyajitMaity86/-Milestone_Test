```python
'''1.13. Write a program for VIBGYOR Spectrum based on their Wavelength using. Wavelength Range:
COLOR            WAVELENGTH (nm)
Violet               400.0-440.0
Indigo               440.0-460.0
Blue                  460.0-500.0
Green               500.0-570.0
Yellow             570.0-590.0
Orange             590.0-620.0
Red                  620.0-720.0'''
def get_color_by_wavelength(wavelength):
    """
    Function to get the color based on the wavelength in the VIBGYOR spectrum.

    Args:
    wavelength (float): The wavelength in nanometers.

    Returns:
    str: The color corresponding to the wavelength.
    """
    if 400.0 <= wavelength < 440.0:
        return "Violet"
    elif 440.0 <= wavelength < 460.0:
        return "Indigo"
    elif 460.0 <= wavelength < 500.0:
        return "Blue"
    elif 500.0 <= wavelength < 570.0:
        return "Green"
    elif 570.0 <= wavelength < 590.0:
        return "Yellow"
    elif 590.0 <= wavelength < 620.0:
        return "Orange"
    elif 620.0 <= wavelength <= 720.0:
        return "Red"
    else:
        return "Wavelength out of VIBGYOR range"

# Test the function with various wavelengths
test_wavelengths = [405.0, 455.0, 485.0, 515.0, 580.0, 600.0, 630.0, 750.0]

print("Wavelength Color Spectrum:")
for wavelength in test_wavelengths:
    color = get_color_by_wavelength(wavelength)
    print(f"Wavelength: {wavelength} nm -> Color: {color}")
```

```python
'''4.Create a simple banking system using object-oriented concepts in Python. Design classes representing different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, and balance inquiry. Utilize inheritance to manage different account types efficiently.
Tasks:
1. Define base class(es) for bank accounts with common attributes and methods.
2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount).
3. Provide methods for deposit, withdraw, and balance inquiry in each subclass.
4. Test the banking system by creating instances of different account types and performing transactions.'''
# Base class for bank accounts
class BankAccount:
    def __init__(self, account_number, account_holder):
        self.__account_number = account_number
        self.__account_holder = account_holder
        self.__balance = 0.0

    # Getter methods
    def get_account_number(self):
        return self.__account_number

    def get_account_holder(self):
        return self.__account_holder

    def get_balance(self):
        return self.__balance

    # Methods for deposit and withdraw
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.__balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount:.2f}. New balance: ${self.__balance:.2f}")
        else:
            print("Withdrawal amount must be positive and not exceed the current balance.")

    # Method for balance inquiry
    def balance_inquiry(self):
        print(f"Account balance: ${self.__balance:.2f}")

# Subclass for savings account
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, interest_rate):
        super().__init__(account_number, account_holder)
        self.__interest_rate = interest_rate

    def apply_interest(self):
        balance = self.get_balance()
        interest = balance * self.__interest_rate / 100
        self.deposit(interest)
        print(f"Applied interest of ${interest:.2f} at rate {self.__interest_rate}%.")

# Subclass for checking account
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, overdraft_limit):
        super().__init__(account_number, account_holder)
        self.__overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > 0 and amount <= self.get_balance() + self.__overdraft_limit:
            new_balance = self.get_balance() - amount
            if new_balance < 0:
                overdraft = -new_balance
                print(f"Overdraft of ${overdraft:.2f} used.")
            super().withdraw(amount)
        else:
            print("Withdrawal amount exceeds balance and overdraft limit.")

# Test the banking system
def test_banking_system():
    # Create a savings account
    savings = SavingsAccount("SA12345", "Alice", 2.5)
    print("\nSavings Account Created:")
    savings.balance_inquiry()
    
    # Perform transactions on savings account
    savings.deposit(1000)
    savings.apply_interest()
    savings.withdraw(500)
    savings.balance_inquiry()
    
    # Create a checking account
    checking = CheckingAccount("CA54321", "Bob", 500)
    print("\nChecking Account Created:")
    checking.balance_inquiry()
    
    # Perform transactions on checking account
    checking.deposit(2000)
    checking.withdraw(2500)
    checking.withdraw(3000)
    checking.balance_inquiry()

# Run the test function
if __name__ == "__main__":
    test_banking_system()
```

```python
'''7. Write a code for Hotel Management System using OOPS:
Create a Room class that has attributes such as room number, room type, rate, and availability (private).
Implement methods to book a room, check in a guest, and check out a guest.
Use encapsulation to hide the room's unique identification number.
Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own specific attributes and methods.'''

class Room:
    _id_counter = 1

    def __init__(self, room_number, room_type, rate):
        self.__id = Room._id_counter
        Room._id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__is_available = True

    # Getter for ID
    def get_id(self):
        return self.__id

    # Method to check availability
    def is_available(self):
        return self.__is_available

    # Method to book a room
    def book_room(self):
        if self.__is_available:
            self.__is_available = False
            print(f"Room {self.room_number} booked successfully.")
        else:
            print(f"Room {self.room_number} is already booked.")

    # Method to check in a guest
    def check_in(self):
        if not self.__is_available:
            print(f"Guest checked in to room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    # Method to check out a guest
    def check_out(self):
        if not self.__is_available:
            self.__is_available = True
            print(f"Guest checked out from room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is already available.")

    # Method to display room details
    def display_info(self):
        availability = "Available" if self.__is_available else "Not Available"
        print(f"Room Number: {self.room_number}")
        print(f"Room Type: {self.room_type}")
        print(f"Rate: ${self.rate:.2f}")
        print(f"Status: {availability}")

# Subclass for SuiteRoom
class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_living_area):
        super().__init__(room_number, "Suite", rate)
        self.has_living_area = has_living_area

    # Overriding display method to include living area information
    def display_info(self):
        super().display_info()
        print(f"Living Area: {'Yes' if self.has_living_area else 'No'}")

# Subclass for StandardRoom
class StandardRoom(Room):
    def __init__(self, room_number, rate, bed_type):
        super().__init__(room_number, "Standard", rate)
        self.bed_type = bed_type

    # Overriding display method to include bed type information
    def display_info(self):
        super().display_info()
        print(f"Bed Type: {self.bed_type}")

# Test the hotel management system
def test_hotel_management_system():
    # Create rooms
    suite1 = SuiteRoom("101", 150.00, True)
    standard1 = StandardRoom("102", 100.00, "Queen")
    
    # Display initial room details
    print("Initial Room Details:")
    suite1.display_info()
    print()
    standard1.display_info()

    # Book rooms
    print("\nBooking Rooms:")
    suite1.book_room()
    standard1.book_room()

    # Check in guests
    print("\nChecking In Guests:")
    suite1.check_in()
    standard1.check_in()

    # Check out guests
    print("\nChecking Out Guests:")
    suite1.check_out()
    standard1.check_out()

    # Display final room details
    print("\nFinal Room Details:")
    suite1.display_info()
    print()
    standard1.display_info()

# Run the test function
if __name__ == "__main__":
    test_hotel_management_system()
```

```python
'''9.Write a code for Event Management System using OOPS:
Create an Event class that has attributes such as name, date, time, location, and list of attendees (private).
Implement methods to create a new event, add or remove attendees, and get the total number of attendees.
Use encapsulation to hide the event's unique identification number.
Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own specific attributes and methods.'''

class Event:
    _id_counter = 1

    def __init__(self, name, date, time, location):
        self.__id = Event._id_counter
        Event._id_counter += 1
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []

    # Getter for ID
    def get_id(self):
        return self.__id

    # Method to add an attendee
    def add_attendee(self, attendee):
        self.__attendees.append(attendee)
        print(f"Attendee {attendee} added to event {self.name}.")

    # Method to remove an attendee
    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Attendee {attendee} removed from event {self.name}.")
        else:
            print(f"Attendee {attendee} not found in event {self.name}.")

    # Method to get the total number of attendees
    def get_total_attendees(self):
        return len(self.__attendees)

    # Method to display event details
    def display_info(self):
        print(f"Event ID: {self.__id}")
        print(f"Name: {self.name}")
        print(f"Date: {self.date}")
        print(f"Time: {self.time}")
        print(f"Location: {self.location}")
        print(f"Total Attendees: {self.get_total_attendees()}")

# Subclass for PrivateEvent
class PrivateEvent(Event):
    def __init__(self, name, date, time, location, invitation_only):
        super().__init__(name, date, time, location)
        self.invitation_only = invitation_only

    # Overriding display method to include invitation-only status
    def display_info(self):
        super().display_info()
        print(f"Invitation Only: {'Yes' if self.invitation_only else 'No'}")

# Subclass for PublicEvent
class PublicEvent(Event):
    def __init__(self, name, date, time, location, public_notice):
        super().__init__(name, date, time, location)
        self.public_notice = public_notice

    # Overriding display method to include public notice information
    def display_info(self):
        super().display_info()
        print(f"Public Notice: {self.public_notice}")

# Test the event management system
def test_event_management_system():
    # Create events
    private_event = PrivateEvent("Birthday Party", "2024-08-01", "18:00", "John's House", True)
    public_event = PublicEvent("Community Meeting", "2024-08-02", "10:00", "Community Hall", "Open to all")

    # Display initial event details
    print("Initial Event Details:")
    private_event.display_info()
    print()
    public_event.display_info()

    # Add attendees
    print("\nAdding Attendees:")
    private_event.add_attendee("Alice")
    private_event.add_attendee("Bob")
    public_event.add_attendee("Charlie")
    public_event.add_attendee("Dave")

    # Remove attendees
    print("\nRemoving Attendees:")
    private_event.remove_attendee("Alice")
    public_event.remove_attendee("Eve")  # Eve is not in the list

    # Display final event details
    print("\nFinal Event Details:")
    private_event.display_info()
    print()
    public_event.display_info()

# Run the test function
if __name__ == "__main__":
    test_event_management_system()
```

```python
'''10. Write a code for Airline Reservation System using OOPS:
Create a Flight class that has attributes such as flight number, departure and arrival airports, departure and arrival times, and available seats (private).
Implement methods to book a seat, cancel a reservation, and get the remaining available seats.
Use encapsulation to hide the flight's unique identification number.
Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their own specific attributes and methods.'''

class Flight:
    _id_counter = 1

    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = Flight._id_counter
        Flight._id_counter += 1
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = total_seats

    # Getter for ID
    def get_id(self):
        return self.__id

    # Method to book a seat
    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked successfully on flight {self.flight_number}.")
        else:
            print(f"No available seats on flight {self.flight_number}.")

    # Method to cancel a reservation
    def cancel_reservation(self):
        self.__available_seats += 1
        print(f"Reservation cancelled successfully on flight {self.flight_number}.")

    # Method to get the remaining available seats
    def get_available_seats(self):
        return self.__available_seats

    # Method to display flight details
    def display_info(self):
        print(f"Flight ID: {self.__id}")
        print(f"Flight Number: {self.flight_number}")
        print(f"Departure Airport: {self.departure_airport}")
        print(f"Arrival Airport: {self.arrival_airport}")
        print(f"Departure Time: {self.departure_time}")
        print(f"Arrival Time: {self.arrival_time}")
        print(f"Available Seats: {self.__available_seats}")

# Subclass for DomesticFlight
class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, domestic_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.domestic_code = domestic_code

    # Overriding display method to include domestic code
    def display_info(self):
        super().display_info()
        print(f"Domestic Code: {self.domestic_code}")

# Subclass for InternationalFlight
class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, international_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.international_code = international_code

    # Overriding display method to include international code
    def display_info(self):
        super().display_info()
        print(f"International Code: {self.international_code}")

# Test the airline reservation system
def test_airline_reservation_system():
    # Create flights
    domestic_flight = DomesticFlight("DF123", "JFK", "LAX", "2024-08-01 08:00", "2024-08-01 11:00", 100, "DOM123")
    international_flight = InternationalFlight("IF456", "JFK", "LHR", "2024-08-01 18:00", "2024-08-02 06:00", 200, "INT456")

    # Display initial flight details
    print("Initial Flight Details:")
    domestic_flight.display_info()
    print()
    international_flight.display_info()

    # Book seats
    print("\nBooking Seats:")
    domestic_flight.book_seat()
    domestic_flight.book_seat()
    international_flight.book_seat()

    # Cancel reservations
    print("\nCancelling Reservations:")
    domestic_flight.cancel_reservation()
    international_flight.cancel_reservation()

    # Display final flight details
    print("\nFinal Flight Details:")
    domestic_flight.display_info()
    print()
    international_flight.display_info()

# Run the test function
if __name__ == "__main__":
    test_airline_reservation_system()
```
